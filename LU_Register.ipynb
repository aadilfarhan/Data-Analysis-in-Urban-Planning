```python
import pandas as pd
```

```python
# Read file from attribute table of block maps
df=pd.read_excel('LU_intersect.xlsx')
```

```python
df.head()
```

```python
# COmbining khasra number, plot number and block number into one column
cols = ['Blk_name', 'Vlg_name', 'Plt_no']
df['survey_no'] = df[cols].apply(lambda row: ' : '.join(row.values.astype(str)), axis=1)
```

```python
# Groupby based in survey number and the maximum area for each revenue number
df_grp= df.groupby(['survey_no'])['Area'].max().reset_index()
```

```python
df_grp.info()
```

```python
# merginf the other details to the dataframe created for maximum area
df_Mahe_final = pd.merge(df_grp, df,  how='left', left_on=['survey_no','Area'], right_on = ['survey_no','Area'])
```

```python
df_Mahe_final.info()
```

```python
# Dropping columns
#df_Mahe_final= df_Mahe_final.drop(columns=['FID', 'FID_Mahe_B', 'FID_ULU_cl', 'Class', 'SHAPE_Leng', 'SHAPE_Area' , 'Code'])
```

```python
df_Mahe_final
```

```python
# Reai=ding excel file for revenue number of roads
df_roads= pd.read_excel('Roads_Merged.xlsx')
```

```python
df_roads.head()
```

```python
cols = ['Blk_name', 'Vlg_name', 'Plt_no']
df_roads['survey_no'] = df_roads[cols].apply(lambda row: ' : '.join(row.values.astype(str)), axis=1)
```

```python
road_survey_no=[]
road_survey_no= df_roads.survey_no.unique()
```

```python
road_survey_no
```

```python
#Naming survey numbers of roads from roads attribute table to roads
road=[]
for road in road_survey_no:
    
    condition = df_Mahe_final['survey_no'] == road
    df_Mahe_final.loc[condition, 'Sub_Class'] = 'Roads'
```

```python
df_Mahe_final.Sub_Class.unique()
```

```python
#Similarly doing for water bodies
df_water_body= pd.read_excel('Yanam_WaterBody.xlsx')
```

```python
df_water_body.head()
```

```python
cols = ['Blk_name', 'Vlg_name', 'Plt_no']
df_water_body['survey_no'] = df_water_body[cols].apply(lambda row: ' : '.join(row.values.astype(str)), axis=1)
```

```python
Water_body_survey_no= df_water_body.survey_no.unique()
```

```python
water_body= []

for water_body in Water_body_survey_no:
    
    condition = df_Mahe_final['survey_no'] == water_body
    df_Mahe_final.loc[condition, 'Sub_Class'] = 'Water_Body'
```

```python
# Saving files into excel
df_Mahe_final.to_excel('Yanam_LU_Register_combined.xlsx')
```
