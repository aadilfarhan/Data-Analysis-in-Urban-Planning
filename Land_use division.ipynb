```python
import pandas as pd
```

```python
#read file (This file is from the attribute table of ULU given by NRSC saved in a csv format) 
df=pd.read_csv('land_use.csv')
```

```python
df.head()
```

```python
#groupng the land use by classes and finding its area
df_landuse= df.groupby('Sub_Class')[['SHAPE_Area']].sum()
```

```python
df_landuse
```

```python
#made the index as default
df_landuse= df_landuse.reset_index()
```

```python
df_landuse.head()
```

```python
#saving the land use classification based on sub class
df_landuse.to_excel('land_use_sub_class_classification.xlsx')
```

```python
#the land use codes as per AMRUT being loaded in a dataframe
df_lu_cl= pd.read_excel('land_use_classification.xlsx')
```

```python
df_lu_cl.head()
```

```python
#Check for whether all the land use classes are there in both data frames
df_landuse['Sub_Class'][~df_landuse['Sub_Class'].isin(df_lu_cl['SUB-CLASS'])]
```

```python
#Renaming column values to match both the data frames
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Agro based & Food Processing', 'Agro based & Food Processing ')
```

```python
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Commercial and Recreational', 'Commercial & Recreational')
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Crematorium / Burial Ground / Grave Yard',
                                                              'Crematorium Burial Groun /Grave Yard')
```

```python
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Other Industries', 'Other Industries ')
```

```python
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Post / Telegraph Office', 'Post/Telegraph Office')
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Scrub land', 'Scrubland')
df_landuse['Sub_Class'] = df_landuse['Sub_Class'].str.replace('Tree Clad Area', 'Tree Clad Area ')
```

```python
df_landuse['Sub_Class'][~df_landuse['Sub_Class'].isin(df_lu_cl['SUB-CLASS'])]
```

```python
df_landuse.head()
```

```python
df_lu_cl.head()
```

```python
# Renaming column names so as to merge the data frames based on sub_class
df_lu_cl=df_lu_cl.rename(columns={"SUB-CLASS": "Sub_Class"})
```

```python
#merging data frames and saving it to df_combined
df_combined= pd.merge(df_landuse,df_lu_cl,on='Sub_Class',how='left')
```

```python
df_combined.head()
```

```python
#groupng the land use by classes and finding its area
df_final= df_combined.groupby('CLASS')[['SHAPE_Area']].sum()
```

```python
df_final
```

```python
#made the index as default
df_final= df_final.reset_index()
```

```python
df_final.to_excel('final_land_use_class.xlsx')
```
